# Polynomial Regression

```python
X = df["zn"]
y = df["medv"]
```

_Here we are initializing two columnsfor two variables for regression part and for splitting also._

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

_Here we are importing the train_test split method for splitting the data. From this we can easily estimate the efficiency of the machine learning algorithms and we can easily compare the results of our own model and the actual result._

```python
p_test = X_test
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
ploy = PolynomialFeatures(3)
```

_Here we are importing the polynomial regression method  and initialzing that method to a variable and also we mentioned the degree of the polynomial. More the degree of the polynomial , the complexity is more. If the complexity is more, then the efficiency is large._

```python
X_train = np.array(X_train).reshape(-1, 1)
X_test = np.array(X_test).reshape(-1, 1)
```

_After splitting the data in to two,we are converting that data into numpy arrray and resizing them so that each value is the list of that array.This helps us to converting each value to a ploynomial regression matrix._

```python
ploy.fit(X_train)
```

```python
X_train = ploy.transform(X_train)
X_test = ploy.transform(X_test)
```

_Here we are polynomially fitting the train data and transfering both train and test data of  'X' variable. This will help us to converting that numpy array to polynomial matrix._

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
rg = LinearRegression()
```

_Polynomial Regression is a complex part of Linear Regression , therefore we ae using Linear Regression also in  Polynomial regression. Here we are importing the linear Regression model and initializing that to a variable._

```python
model = rg.fit(X_train,y_train)
```

_here we are linearing fitting the data of train values in x variable so that it calculates the values of the parameters._

```python
model.coef_
```

```python
model.intercept_
```

_If want the coefficient and intercept  of the model then we can use above two code lines. it returns the coefficient and intercept values as a list of numpy array._

```python
y_predict = model.predict(X_test)
```

_here we are predicting the 'y' values using the inbuilt predict method. this helps us to plot the graph._

```python
plt.scatter(X_test[:, 1], y_predict, c = "b")
plt.scatter(X_test[:, 1], y_test, c = "g")
plt.show()
```

_Here we are plotting two graphs with X axis as the test data of 'x' variable and y axis as prdiction of y and test data of the 'y' variable. This shows least error and that passes through maximum points._

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

_Here we are importing the mean_squared error , mean_absolute_error and r2_score module using the sklearn.This will help us to calculate dirctly all the errors by mentioning their names only._

```python
mean_absolute_error(y_test,y_predict)
```

```python
mean_squared_error(y_test,y_predict)
```

```python
r2_score(y_test, y_predict)
```
